```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("./Sales_Data/Sales_April_2019.csv")
files = [file for file in os.listdir('./Sales_Data')]
all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Sales_Data/" + file)
    all_months_data = pd.concat([all_months_data, df])

all_months_data.to_csv("all_data.csv", index=False)
```

```python
alldata=all_months_data
alldata
```

```python
alldata.isnull().sum()
```

# Clean NaN value 

```python
nan=alldata[alldata.isna().any(axis=1)]
nan.head()
```

```python
alldata = alldata.dropna(how='any')
```

```python
alldata.head()
```

```python
##Find or and delete
alldata=alldata[alldata['Order Date'].str[0:2] !='Or']
                 
alldata.head()
alldata.info()
```

# Convert Column into proper data type

```python
alldata['Quantity Ordered'] = pd.to_numeric(alldata['Quantity Ordered'])
alldata['Price Each'] = pd.to_numeric(alldata['Price Each'])
alldata.head()
alldata.info()
```

# Augment data with additional columns

```python
alldata['Month']=alldata['Order Date'].str[0:2]
alldata['Month']= alldata['Month'].astype('int')
alldata
```

# Adding the sales column

```python
alldata['Sales']=alldata['Quantity Ordered']*alldata['Price Each']
alldata.head()
```

# Add City column

```python
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]



alldata['City']=alldata['Purchase Address'].apply(lambda x: get_city(x)+', ('+get_state(x)+')')
alldata.head()
```

# Question 1:What was the best month for sales?How much was warned at that months?

```python
alldata.groupby('Month').sum()['Sales']
```

```python
results=alldata.groupby('Month').sum()
```

```python
month=range(1,13)
plt.bar(month, results['Sales'])
plt.xticks(month)
plt.ylabel("Sales in USD $(in milliion)" )
plt.xlabel("Months of sales")
plt.title("Sales Data")
plt.show()
```

# Most item sold in cities

```python
results=alldata.groupby('City').sum()
```

```python
# cities=alldata['City'].unique()

cities=[city for city ,df in alldata.groupby('City')]
plt.bar(cities, results['Sales'])
plt.xticks(cities,rotation='vertical',size=8)
plt.ylabel("Sales in USD $(in milliion)" )
plt.xlabel("City Name")
plt.title("Sales Data")
plt.show()
```

```python
alldata['Hour']=alldata['Order Date'].dt.hour
alldata['Minute']=alldata['Order Date'].dt.minute
alldata['Count']=1
alldata.head()
```

```python
alldata= alldata.drop(columns='Purchase Address',axis=1)
alldata
```

```python
alldata['Order Date']= pd.to_datetime(alldata['Order Date'])
```

# Orders per hour


```python
hours=[hours for hours,df in alldata.groupby('Hour')]
plt.plot(hours,alldata.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel("Hours")
plt.ylabel("No. of Orders")
plt.grid()
plt.show()
```

# Which item is most often sold together?

```python
df=alldata[alldata['Order ID'].duplicated(keep=False)]
df.head(20)
```

# Grouping items into same column having same Order Id

```python
df['Grouped']=df.groupby('Order ID')['Product'].transform(lambda x:', '.join(x))
df=df[['Order ID','Grouped']].drop_duplicates()
df.head()
```

# Most sold item/products

```python
alldata.head()
```

```python
product_data=alldata.groupby('Product')
quantity_ordered=product_data.sum()['Quantity Ordered']
products=[product for product ,df in product_data]
plt.bar(products,quantity_ordered,color='#80393C')
plt.xticks(products,rotation='vertical',size=5)
plt.xlabel("Product name")
plt.ylabel("Quantity Ordered")
plt.title("Product vs Quantity_Ordered")
plt.show()
```

```python
prices=alldata.groupby('Product').mean()['Price Each']
```

# Adding the 2 labels in y axis in same bar garph

```python
fig,ax1=plt.subplots()
ax2=ax1.twinx()
ax1.bar(products,quantity_ordered)
ax2.plot(products,prices,'-b')
ax1.set_xlabel('Product Name',color='b')
ax1.set_ylabel('Quantity Ordred',color='r')
ax2.set_ylabel('Prices in $',color='g')
ax1.set_xticklabels(products,rotation='vertical',size=8)
plt.show()



```
